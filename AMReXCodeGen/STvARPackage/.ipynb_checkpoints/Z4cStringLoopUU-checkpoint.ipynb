# Metrics

```python
gamtilde_UU = inversemetric(gamtilde_LL)
RHSString += gamtilde_UU.symb2expr()
```

```python
gam_LL = stvarrank2('gam_LL', sym = 'sym01')
gam_LL.expr = (chi.symb**(-1))*gamtilde_LL.symb

gam_UU = stvarrank2('invgam_UU', sym = 'sym01')
gam_UU.expr = chi.symb*gamtilde_UU.symb

RHSString += gam_LL.symb2expr()
RHSString += gam_UU.symb2expr()
```

```python
n_L = stvarrank1('n_L', dim = 4)
n_L.expr[0] += -alpha.symb

n_U = stvarrank1('n_U', dim = 4)
n_U.expr[0] += 1/alpha.symb
for i in range(3):
    n_U.expr[i+1] += -beta_U.symb[i]/alpha.symb
    
RHSString += n_L.symb2expr()
RHSString += n_U.symb2expr()
```

```python
beta_L = stvarrank1('beta_L')
for i in range(3):
    for j in range(3):
        beta_L.expr[i] += gam_LL.symb[i][j]*beta_U.symb[j]
        
RHSString += beta_L.symb2expr()
```

```python
g4_LL = stvarrank2('g4_LL', sym = 'sym01', dim = 4)
g4_LL.expr[0][0] = -alpha.symb**2
for i in range(3):
    g4_LL.expr[0][0] += beta_U.symb[i]*beta_L.symb[i]
    g4_LL.expr[0][i+1] += beta_L.symb[i]
    g4_LL.expr[i+1][0] += beta_L.symb[i]
    
for i in range(3):
    for j in range(3):
        g4_LL.expr[i+1][j+1] += gam_LL.symb[i][j]
        
RHSString += g4_LL.symb2expr()
```

```python
g4_UU = stvarrank2('g4_UU', sym = 'sym01', dim = 4)
g4_UU.expr[0][0] = -alpha.symb**(-2)
for i in range(3):
    g4_UU.expr[0][i+1] += alpha.symb**(-2)*beta_U.symb[i]
    g4_UU.expr[i+1][0] += alpha.symb**(-2)*beta_U.symb[i]
    
for i in range(3):
    for j in range(3):
        g4_UU.expr[i+1][j+1] += gam_UU.symb[i][j]-alpha.symb**(-2)*beta_U.symb[i]*beta_U.symb[j]
        
RHSString += g4_UU.symb2expr()
```

```python
gam4_LL = stvarrank2('gam4_LL', sym = 'sym01', dim = 4)
for a in range(4):
    for b in range(4):
        gam4_LL.expr[a][b] += g4_LL.symb[a][b] + n_L.symb[a]*n_L.symb[b]
        
RHSString += gam4_LL.symb2expr()
```

```python
gam4_UU = stvarrank2('gam4_UU', sym = 'sym01', dim = 4)
for a in range(4):
    for b in range(4):
        gam4_UU.expr[a][b] += g4_UU.symb[a][b] + n_U.symb[a]*n_U.symb[b]
        
RHSString += gam4_UU.symb2expr()
```

```python
gam4_UL = stvarrank2('gam4_UL', dim = 4)
for a in range(4):
    for b in range(4):
        gam4_UL.expr[a][b] += sp.eye(4)[a,b] + n_U.symb[a]*n_L.symb[b]
        
RHSString += gam4_UL.symb2expr()        
```

```python
gam4_LU = stvarrank2('gam4_LU', dim = 4)
for a in range(4):
    for b in range(4):
        gam4_LU.expr[a][b] += sp.eye(4)[a,b] + n_L.symb[a]*n_U.symb[b]
        
RHSString += gam4_LU.symb2expr()
```

```python
a_L = stvarrank1('a_L', dim = 4, isreal = True)
a_L.expr[0] +=  - alpha.symb*A.symb
for i in range (3):
    a_L.expr[0] += beta_U.symb[i]*A_L.symb[i]
    a_L.expr[i+1] += A_L.symb[i]
    
RHSString += a_L.symb2expr()
```

```python
a_U = stvarrank1('a_U', dim = 4, isreal = True)
for a in range(4):
    for b in range(4):
        a_U.expr[a] += g4_UU.symb[a][b]*a_L.symb[b]
        
RHSString += a_U.symb2expr()
```

```python
CovDtildedA_LLL = stvarrank3('CovDtildedA_LLL')
for k in range(3):
    for j in range(3):
        for l in range(3):
            CovDtildedA_LLL.expr[l][j][k] += dDDA_LLL.symb[l][j][k]
            for q in range(3):
                CovDtildedA_LLL.expr[l][j][k] += -Gamtilde_ULL.symb[q][k][j]*dDA_LL.symb[l][q] - Gamtilde_ULL.symb[q][k][l]*dDA_LL.symb[q][j]
                
RHSString += CovDtildedA_LLL.symb2expr()
```

```python
RHS_E_U = stvarrank1('E_U', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    RHS_E_U.expr[i] += alpha.symb*Ksclr.symb*E_U.symb[i] + AdvDE_U.symb[i] + dKODFullE_U.symb[i]
    RHS_E_U.expr[i] += -alpha.symb*echarge.symb*scrJ_U.symb[i]
    
    for j in range(3):
        RHS_E_U.expr[i] += alpha.symb*chi.symb*gamtilde_UU.symb[i][j]*dDZ_L.symb[j]
        RHS_E_U.expr[i] += -E_U.symb[j]*dDbeta_UL.symb[i][j]
        
        for k in range(3):
            for l in range(3):
                RHS_E_U.expr[i] += gam_UU.symb[i][j]*gam_UU.symb[k][l]*dDalpha_L.symb[l]*(dDA_LL.symb[k][j]-dDA_LL.symb[j][k])
                RHS_E_U.expr[i] += alpha.symb*gam_UU.symb[i][j]*gam_UU.symb[k][l]*(CovDtildedA_LLL.symb[l][j][k]-CovDtildedA_LLL.symb[j][l][k])
                RHS_E_U.expr[i] += (alpha.symb/2)*chi.symb*gamtilde_UU.symb[i][j]*gamtilde_UU.symb[k][l]*(dDA_LL.symb[l][j]*dDchi_L.symb[k] - dDA_LL.symb[j][k]*dDchi_L.symb[l])

RHSString += RHS_E_U.setisymb()
```

```python
RHS_A = stvar('A', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_A.expr += alpha.symb*Ksclr.symb*A.symb - alpha.symb*Z.symb + AdvDA.symb + dKODFullA.symb
for i in range(3):
    RHS_A.expr += alpha.symb*chi.symb*A_L.symb[i]*Gamtilde_U.symb[i]
    for j in range(3):
        RHS_A.expr += -alpha.symb*gam_UU.symb[i][j]*dDA_LL.symb[i][j] + (alpha.symb/2)*A_L.symb[i]*gamtilde_UU.symb[i][j]*dDchi_L.symb[j]
        RHS_A.expr += -gam_UU.symb[i][j]*A_L.symb[i]*dDalpha_L.symb[j]
        
RHSString += RHS_A.setisymb()
```

```python
RHS_A_L = stvarrank1('A_L', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
for i in range(3):
    RHS_A_L.expr[i] += -alpha.symb*dDA_L.symb[i] - A.symb*dDalpha_L.symb[i] + AdvDA_L.symb[i] + dKODFullA_L.symb[i]
    for j in range(3):
        RHS_A_L.expr[i] += -(alpha.symb/chi.symb)*gamtilde_LL.symb[i][j]*E_U.symb[j] + dDbeta_UL.symb[j][i]*A_L.symb[j]
        
RHSString += RHS_A_L.setisymb()
```

```python
RHS_Z = stvar('Z', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_Z.expr += -alpha.symb*echarge.symb*scrJ.symb - alpha.symb*kappaEM.symb*Z.symb + AdvDZ.symb + dKODFullZ.symb

for i in range(3):
    RHS_Z.expr += alpha.symb*dDE_UL.symb[i][i]-sp.Rational(3,2)*(alpha.symb/chi.symb)*E_U.symb[i]*dDchi_L.symb[i]
    
RHSString += RHS_Z.setisymb()
```

```python
A_U = stvarrank1('A_U')
for i in range(3):
    for j in range(3):
        A_U.expr[i] += gam_UU.symb[i][j]*A_L.symb[j]
        
RHSString += A_U.symb2expr()
```

```python
RHS_Pi1 = stvar('Pi1', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_Pi1.expr += AdvDPi1.symb + dKODFullPi1.symb + alpha.symb*(Ksclr.symb*Pi1.symb - dVdphi1.symb)
#RHS_Pi1.expr += 2*alpha.symb*echarge.symb*(A.symb/alpha.symb)*dtphi2.symb
for i in range(3):
    #RHS_Pi1.expr += 2*alpha.symb*echarge.symb*A_U.symb[i]*dDphi2_L.symb[i]
    
    for j in range(3):
        RHS_Pi1.expr += alpha.symb*gam_UU.symb[i][j]*dDDphi1_LL.symb[i][j]
        for k in range(3):
            RHS_Pi1.expr += -alpha.symb*gam_UU.symb[i][j]*Gam_ULL.symb[k][i][j]*dDphi1_L.symb[k]

for mu in range(4):
    RHS_Pi1.expr += -alpha.symb*echarge.symb*echarge.symb*a_U.symb[mu]*a_L.symb[mu]*phi1.symb
    RHS_Pi1.expr += 2*alpha.symb*echarge.symb*a_U.symb[mu]*Dphi2_L.symb[mu]
    
RHSString += RHS_Pi1.setisymb()
```

```python
RHS_Pi2 = stvar('Pi2', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_Pi2.expr += AdvDPi2.symb + dKODFullPi2.symb + alpha.symb*(Ksclr.symb*Pi2.symb - dVdphi2.symb)
#RHS_Pi2.expr += -2*alpha.symb*echarge.symb*(A.symb/alpha.symb)*dtphi1.symb
for i in range(3):
    #RHS_Pi2.expr += -2*alpha.symb*echarge.symb*A_U.symb[i]*dDphi1_L.symb[i]
    
    for j in range(3):
        RHS_Pi2.expr += alpha.symb*gam_UU.symb[i][j]*dDDphi2_LL.symb[i][j]
        for k in range(3):
            RHS_Pi2.expr += -alpha.symb*gam_UU.symb[i][j]*Gam_ULL.symb[k][i][j]*dDphi2_L.symb[k]

for mu in range(4):
    RHS_Pi2.expr += -alpha.symb*echarge.symb*echarge.symb*a_U.symb[mu]*a_L.symb[mu]*phi2.symb
    RHS_Pi2.expr += -2*alpha.symb*echarge.symb*a_U.symb[mu]*Dphi1_L.symb[mu]
    

RHSString += RHS_Pi2.setisymb()
```

```python
RHS_delta1 = stvar('delta1', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_delta1.expr += AdvDdelta1.symb + dKODFulldelta1.symb + alpha.symb*(Ksclr.symb*delta1.symb - dVdsigma1.symb)
#RHS_Pi1.expr += 2*alpha.symb*echarge.symb*(A.symb/alpha.symb)*dtphi2.symb
for i in range(3):
    #RHS_Pi1.expr += 2*alpha.symb*echarge.symb*A_U.symb[i]*dDphi2_L.symb[i]
    
    for j in range(3):
        RHS_delta1.expr += alpha.symb*gam_UU.symb[i][j]*dDDsigma1_LL.symb[i][j]
        for k in range(3):
            RHS_delta1.expr += -alpha.symb*gam_UU.symb[i][j]*Gam_ULL.symb[k][i][j]*dDsigma1_L.symb[k]
    
RHSString += RHS_delta1.setisymb()
```

```python
RHS_delta2 = stvar('delta2', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_delta2.expr += AdvDdelta2.symb + dKODFulldelta2.symb + alpha.symb*(Ksclr.symb*delta2.symb - dVdsigma2.symb)
#RHS_Pi1.expr += 2*alpha.symb*echarge.symb*(A.symb/alpha.symb)*dtphi2.symb
for i in range(3):
    #RHS_Pi1.expr += 2*alpha.symb*echarge.symb*A_U.symb[i]*dDphi2_L.symb[i]
    
    for j in range(3):
        RHS_delta2.expr += alpha.symb*gam_UU.symb[i][j]*dDDsigma1_LL.symb[i][j]
        for k in range(3):
            RHS_delta2.expr += -alpha.symb*gam_UU.symb[i][j]*Gam_ULL.symb[k][i][j]*dDsigma1_L.symb[k]
    
RHSString += RHS_delta2.setisymb()
```
